###### The Urban population as percentage of total population and Economy are interdependent, its clear from the summary statistics;

Population and economy are interconnected, and the relationship between them can be complex. Both factors can influence each other, and the direction of the relationship can vary based on different circumstances and contexts. Here are a few points to consider:

###### Economy Drives Population Growth: 
In regions or countries experiencing economic growth, there might be an increase in job opportunities, higher incomes, and improved living standards. This can lead to higher birth rates and population growth as families feel more secure about their economic prospects.

###### Population Drives Economic Growth:
A growing population can contribute to economic growth by providing a larger labor force, potentially leading to increased productivity and innovation. A larger consumer base can also stimulate demand for goods and services, fostering economic development.

```python
##subseting our dataset into Countries betwee 4 -716 economy

CountryEconomy = UrbanPop[(UrbanPop["Economy"] >= 4) & (UrbanPop["Economy"] <= 716)]
CountryEconomy
```

```python
##Population prediction from 2025 t0 2050 per country
Country25_50 = CountryEconomy[(CountryEconomy['Year'] >= 2025)]
Country25_50 
```

```python
CountryPop = Country25_50.groupby(["Economy_Label"])[['Absolute_value_in_thousands']].sum().reset_index()
CountryPop = CountryPop.sort_values(by=('Absolute_value_in_thousands'), ascending=False)
```

```python
ax = sns.barplot(x="Economy_Label", data=CountryPop.head(10), y= 'Absolute_value_in_thousands')
ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha="right")
plt.title("Top ten Countries with highest population prediction from 2025 t0 2050")
plt.tight_layout()
plt.show()
```

```python
CountryPop = Country25_50.groupby(["Economy_Label"])[['Absolute_value_in_thousands']].sum().reset_index()
CountryPop = CountryPop.sort_values(by=('Absolute_value_in_thousands'), ascending=False)
ax = sns.barplot(x="Economy_Label", data=CountryPop.tail(10), y= 'Absolute_value_in_thousands')
ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha="right")
plt.title("Top ten countries with lowest population prediction from 2025 t0 2050")
plt.tight_layout()
plt.show()
```

#### Observation
The Asian countries namely India, China, and Pakistan are expected to record the largest population come 2050, in Africa Nigeria, Ethiopia, and DRC will record the largest population respectively
